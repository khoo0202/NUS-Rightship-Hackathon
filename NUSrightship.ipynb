```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("vessel_movements_PPT.csv")
```

```python
df
```

```python
column_headers = list(df.columns)
column_headers
```

```python
df.sort_values(by="timestamp",ascending = False, inplace = True)
df
```

```python
df["moving"] = False
```

```python
navigational_status = [0, 3, 4, 8, 11, 12]

df.loc[(df["speed"] > 1) | (df["nav_stat"].isin(navigational_status)), "moving"] = True
df
```

```python
df["operating_mode"] = None
```

```python
df.loc[((df["moving"]==True) & (df["maneuvering_zone"].isna())),"operating_mode"]= "Transit"
df.loc[((df["moving"]==True) & (df["maneuvering_zone"].notna())),"operating_mode"]= "Maneuvering"
df.loc[((df["speed"] < 1) & (df["anchorage"].notna())), "operating_mode"]="Anchorage"
df.loc[((df["speed"] < 1) & (df["berth"].notna())), "operating_mode"]="berth"
df  #Step 1 done
```

```python
timesort = df.sort_values(by=['vessel_id','timestamp'],ascending = [True,True])
timesort.loc[:,["vessel_id","speed","operating_mode","maneuvering_zone","anchorage","berth"]].head(20)
```
